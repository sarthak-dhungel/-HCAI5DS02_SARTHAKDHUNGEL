```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import fetch_california_housing
```

Load the dataset as a pandas DataFrame

```python
data = fetch_california_housing(as_frame=True)
df = data.frame
```

```python
# Display first 5 rows
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# Display basic info about the dataset
print("\nDataset Info:")
print(df.info())
```

```python
# Display statistical summary
print("\nStatistical Summary:")
print(df.describe())
```

# **Problem 1 â€“ Sorting**

```python
# 1. Create a DataFrame with only the 'MedInc' column
med_income = df[['MedInc']]
print("First 5 rows of Median Income:")
print(med_income.head())
```

```python
# 2. Create a DataFrame with 'Population' and 'Latitude' columns
pop_lat = df[['Population', 'Latitude']]
print("\nFirst 5 rows of Population and Latitude:")
print(pop_lat.head())
```

```python
# 3. Create a DataFrame with 'HouseAge' and 'AveRooms' columns
house_age_rooms = df[['HouseAge', 'AveRooms']]
print("\nFirst 5 rows of House Age and Average Rooms:")
print(house_age_rooms.head())
```

# A. Subsetting Rows

```python
# 1. Filter where MedInc > 8.0
high_income = df[df['MedInc'] > 8.0]
print("High Income Houses (MedInc > 8.0):")
print(high_income.head())
```

```python
# 2. Filter where Latitude > 37
north_california = df[df['Latitude'] > 37]
print("\nHouses in North California (Latitude > 37):")
print(north_california.head())
```

```python
# 3. Filter where AveRooms > 6.0 and AveOccup < 2.0
spacious_low_occupancy = df[(df['AveRooms'] > 6.0) & (df['AveOccup'] < 2.0)]
print("\nSpacious Low Occupancy Houses (AveRooms > 6.0 and AveOccup < 2.0):")
print(spacious_low_occupancy.head())
```

# B. Subsetting Categorical Equivalents

```python
# 1. Create 'Region' column based on Latitude
def categorize_region(lat):
    if lat > 37:
        return 'North'
    elif lat > 35:
        return 'Central'
    else:
        return 'South'

df['Region'] = df['Latitude'].apply(categorize_region)
```

```python
# 2. Filter where Region is 'North' or 'Central'
north_central_region = df[df['Region'].isin(['North', 'Central'])]
print("\nHouses in North or Central Region:")
print(north_central_region.head())
```

# Problem 3 â€“ Exploratory Data Analysis (EDA)

```python
# A. Value per room analysis

# 1. Create new column
df['value_per_room'] = df['MedHouseVal'] / df['AveRooms']

# 2. Filter where value_per_room > 1
high_vpr = df[df['value_per_room'] > 1]

# 3. Sort by descending value_per_room
high_vpr_sorted = high_vpr.sort_values(by='value_per_room', ascending=False)

# 4. Display top 5 rows
print("Top 5 houses with highest value per room:")
print(high_vpr_sorted[['MedHouseVal', 'AveRooms', 'value_per_room']].head())
```

B. Q2: In high population areas, who has the highest income per person?

```python
# B. Income per person in high population areas

# 1. Create income per person column
df['income_per_person'] = df['MedInc'] / df['Population']

# 2. Filter where Population > 5000
dense_areas = df[df['Population'] > 5000]

# 3. Sort by descending income_per_person
rich_dense_areas = dense_areas.sort_values(by='income_per_person', ascending=False)

# 4. Display top 5 rows
print("\nTop 5 high population areas with highest income per person:")
print(rich_dense_areas[['MedInc', 'Population', 'income_per_person']].head())
```

# Problem 4 â€“ GroupBy Exercises

```python
# A. Percentage of total house value by Region

# 1. Total house value
total_house_value = df['MedHouseVal'].sum()

# 2. Sum of MedHouseVal per region
region_value_sum = df.groupby('Region')['MedHouseVal'].sum()

# 3. Percentage share
region_value_percent = (region_value_sum / total_house_value) * 100

print("Percentage of total house value from each Region:")
print(region_value_percent)
```

B. Q2: What percent of total houses belong to different age groups?

```python
# B. Age group distribution

# 1. Define AgeGroup column
def assign_age_group(age):
    if age < 20:
        return 'New'
    elif age < 40:
        return 'Mid'
    else:
        return 'Old'

df['AgeGroup'] = df['HouseAge'].apply(assign_age_group)

# 2. Total houses
total_houses = len(df)

# 3. Count per age group
age_group_counts = df['AgeGroup'].value_counts()

# 4. Percentage share
age_group_percent = (age_group_counts / total_houses) * 100

print("\nPercentage of houses by Age Group:")
print(age_group_percent)
```

Advanced Exercises â€“ Correlation Analysis & Missing Data Handling

```python
# A. Pearson correlation with MedHouseVal

# Calculate correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Extract correlations with MedHouseVal
correlations_with_target = correlation_matrix['MedHouseVal'].sort_values(ascending=False)

print("Correlation of features with MedHouseVal:")
print(correlations_with_target)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Backup original column
df['AveRooms_original'] = df['AveRooms'].copy()

# 1. Introduce 5% NaNs in AveRooms
np.random.seed(42)  # for reproducibility
missing_indices = df.sample(frac=0.05).index
df.loc[missing_indices, 'AveRooms'] = np.nan

print(f"\nNumber of missing values in AveRooms after simulation: {df['AveRooms'].isna().sum()}")

# 2. Impute with median
median_value = df['AveRooms'].median()
df['AveRooms'] = df['AveRooms'].fillna(median_value)

print(f"Missing values after imputation: {df['AveRooms'].isna().sum()}")

# 3. Visualize before and after imputation
plt.figure(figsize=(12, 5))

# Before
plt.subplot(1, 2, 1)
sns.histplot(df['AveRooms_original'], kde=True, bins=30, color='skyblue')
plt.title("AveRooms - Original Distribution")

# After
plt.subplot(1, 2, 2)
sns.histplot(df['AveRooms'], kde=True, bins=30, color='lightgreen')
plt.title("AveRooms - After Median Imputation")

plt.tight_layout()
plt.show()
```

2. Vectorization vs Loops â€“ Time Comparison

```python
import numpy as np
import time

# Sample array
arr = np.arange(1_000_000)

# 1. Square using a for-loop
def square_with_loop(array):
    result = []
    for x in array:
        result.append(x ** 2)
    return result

# 2. Square using NumPy vectorization
def square_with_numpy(array):
    return array ** 2

# Timing the loop
start_time = time.time()
square_with_loop(arr)
loop_time = time.time() - start_time

# Timing the vectorized version
start_time = time.time()
square_with_numpy(arr)
vector_time = time.time() - start_time

print(f"Time taken with loop: {loop_time:.4f} seconds")
print(f"Time taken with NumPy vectorization: {vector_time:.4f} seconds")
```
